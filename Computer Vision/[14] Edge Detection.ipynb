# Edge Detection

```python
import numpy as np
import cv2
import warnings
import matplotlib.pyplot as plt

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/dog_1.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Canny Edge

```python
# canny edge
edges = cv2.Canny(image = img, threshold1 = 127, threshold2 = 127)
plt.imshow(edges)
```

### Tweak

```python
# find median pixel value
med_val = np.median(img)
med_val
```

Apply standard formula to set better lower and upper threshold

```python
# lower = either 0 or 30% of med value
# upper = either 255 or 130% of med value
lower = int(max(0, 0.7 * med_val))
upper = int(min(255, 1.3 * med_val))
```

```python
edges = cv2.Canny(image = img, threshold1 = lower, threshold2 = upper + 10)
plt.imshow(edges)
```

### Blur image and Canny

```python
# blur image!! and then apply Canny edge detection (IN THIS IMAGE NOT WORKS WELL!!)
blurred_img = cv2.blur(img, ksize = (5,5))
edges = cv2.Canny(image = blurred_img, threshold1 = lower, threshold2 = upper)
plt.imshow(edges)
```
