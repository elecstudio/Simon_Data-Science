# Gradient

```python
import numpy as np
import cv2
import warnings
import matplotlib.pyplot as plt

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/bricks.png", 0)
show_pic(img)
```

## Calculate x-gradient SOBEL

```python
# [1] calculate X-gradient sobel ( ksize = 5)
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 5)
show_pic(sobelX)
```

## Calculate y-gradient SOBEL

```python
# [2] calculate Y -gradient sobel ( ksize = 5)
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 5)
show_pic(sobelY)
```

## Blend images

```python
# Blend images using addWeighted
blended = cv2.addWeighted(src1 = sobelX, src2 = sobelY, alpha = 0.5, beta = 0.5, gamma = 0)
show_pic(blended)
```

## Morphological Gradient

```python
kernel = np.ones((5,5), dtype = np.uint8)
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
show_pic(gradient)
```
