# Threshold

```python
import numpy as np
import cv2
import warnings
import matplotlib.pyplot as plt

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/rainbow.jpg", 0)
plt.imshow(img, cmap = 'gray')
```

## Use Threshold

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh1, cmap = 'gray') # below 127 -> 0 and above 127 -> keep same
```

## Binary Threshold

```python
# open a new img
img = cv2.imread("Computer-Vision-with-Python/DATA/crossword.jpg", 0)
```

```python
ret, thresh1 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY)
show_pic(thresh1)
```

# Adaptive Threshold

```python
# adaptive threshold
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
show_pic(th2)
```

# Adaptive Threshold with OTSU method

```python
ret, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
show_pic(thresh)
```
