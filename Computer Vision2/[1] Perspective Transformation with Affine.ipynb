# Perspective Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("DATA/tilted_diary.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
point_a = np.float32([[440,50],[830,270],[210,590],[610,790]])
point_b = np.float32([[0,0], [400,0], [0,500],[400,500]])
```

```python
M = cv2.getPerspectiveTransform(point_a, point_b)
```

```python
wraped = cv2.warpPerspective(img, M, (400,500))
```

```python
plt.imshow(wraped)
```
