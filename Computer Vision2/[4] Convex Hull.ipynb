# Convex Hull

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread("DATA/hand.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

## Threshold

```python
ret, thresh = cv2.threshold(gray_img, 180, 255, 0)
show_pic(thresh)
```

## Contours

```python
image, contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
```

```python
blank = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)
```

If our image is in white background the entire frame is regarded as a contour and need to remove. We calculate contour area and sort them and exclude the largest contour area which is the exterior box.

```python
n = len(contours) - 2
contours = sorted(contours, key = cv2.contourArea, reverse = False)[:n]

for c in contours:
    hull = cv2.convexHull(c)
    cv2.drawContours(img, [hull], 0, (0,0,255), 8)
```
