# Contour Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread("DATA/chair2.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

## Canny Edge Detection to remove noises

```python
canny_img = cv2.Canny(gray_img, 130, 190)
show_pic(canny_img)
```

## Find contours

```python
image, contours, hierarchy = cv2.findContours(canny_img.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
```

## Draw contours

```python
blank = np.zeros(img.shape, dtype = np.uint8)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(blank, contours, i, (255,0,0), 4)
    else:
        cv2.drawContours(blank, contours, i, (0,0,255), 4)
plt.imshow(blank)
```
