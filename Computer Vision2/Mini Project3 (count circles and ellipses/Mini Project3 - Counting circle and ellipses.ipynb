# Mini Project3 - Counting circle and ellipses

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("DATA2/images/blobs.jpg", 0)
plt.figure(figsize = (10,10))
plt.imshow(img, cmap = 'gray')
```

## Create a blob detector object

```python
# blob detector with default parameter
detector = cv2.SimpleBlobDetector_create()
```

## Detect Key points

```python
key_points = detector.detect(img)
key_points
# draw blobs on the image
blank = np.zeros((2,2))
blobs = cv2.drawKeypoints(img, key_points, blank, (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(blobs)
```

## Put the number of blobs found on image

```python
num_blobs = len(key_points)
text = "Total number of blobs : "+str(num_blobs)
cv2.putText(blobs, text, (20,550), cv2.FONT_HERSHEY_SIMPLEX, 1, (100,0,255), 2)
plt.figure(figsize = (10,10))
plt.imshow(blobs)
```

## Distingiush circle and ellipse using parameters

```python
# set filtering parameter
params = cv2.SimpleBlobDetector_Params()

# set area param filtering
params.filterByArea = True
params.minArea = 100

# set circularity filtering param
params.filterByCircularity = True
params.minCircularity = 0.9

# set convexity filtering param
params.filterByConvexity = False
params.minConvexity = 0.2

# set inertia filtering param
params.filterByInertia = True
params.minInertiaRatio = 0.01

# create a detector with params
detector = cv2.SimpleBlobDetector_create(params)
```

## Try again

```python
key_points = detector.detect(img)
key_points
# draw blobs on the image
blank = np.zeros((2,2))
blobs = cv2.drawKeypoints(img, key_points, blank, (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
num_blobs = len(key_points)
text = "Total number of circles : "+str(num_blobs)
cv2.putText(blobs, text, (20,550), cv2.FONT_HERSHEY_SIMPLEX, 1, (100,100,255), 2)
plt.figure(figsize = (10,10))
plt.imshow(blobs)
```
