# Mini Project4 - Finding Waldo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("DATA2/images/WaldoBeach.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (10,10))
plt.imshow(img)
```

## Load template image

```python
template = cv2.imread("DATA2/images/waldo.jpg", 0)
```

## Use templateMatching

```python
result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

## Create bounding box

```python
top_left = max_loc
waldo_shape = template.shape[:2]
bottom_right = (top_left[0] + waldo_shape[0], top_left[1] + waldo_shape[1])
cv2.rectangle(img, top_left, bottom_right, (0,0,255), 5)

plt.figure(figsize = (10,10))
plt.imshow(img)
```

## Heatmap for template matching

```python
plt.figure(figsize = (10,10))
plt.imshow(result)
```
