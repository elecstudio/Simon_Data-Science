# Contour Sorting

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = cv2.imread("DATA/shapes2.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show_pic(gray_img)
```

```python
canny_img = cv2.Canny(gray_img, 33, 256)
# show_pic(canny_img)
```

```python
image, contours, hierarchy = cv2.findContours(canny_img.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
blank = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)
```

```python
cv2.drawContours(blank, contours, -1, (255,0,0), 4)
plt.imshow(blank)
```

## Sort by size

```python
def get_contour_area(contours):
    all_areas = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        all_areas.append(area)
        
    return all_areas
```

```python
# sort contours by area and calculate area
sorted_areas = sorted(contours, key = cv2.contourArea, reverse = True)
areas = get_contour_area(contours)
```

```python
# see the result [Largest]
cv2.drawContours(img, sorted_areas[0], -1, [255,0,0], 8)
plt.imshow(img)
```

```python
# Second Largest
cv2.drawContours(img, sorted_areas[1], -1, [0,255,0], 8)
plt.imshow(img)
```

```python
# Third Largest
cv2.drawContours(img, sorted_areas[2], -1, [0,0,255], 10)
plt.imshow(img)
```
