# Blob detection (detect sunflowers)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("DATA/sunflowers.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

## Create a detector

```python
detector = cv2.SimpleBlobDetector_create()
```

## Detect key points

```python
key_points = detector.detect(img_gray)
```

## Draw detected key points on image

```python
blank = np.zeros((1,1))
blobs = cv2.drawKeypoints(img_gray, key_points, blank, (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(blobs)
```
